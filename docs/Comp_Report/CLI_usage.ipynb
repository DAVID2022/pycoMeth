# Meth_Comp CLI usage

## Activate virtual environment

```bash
# Using virtualenvwrapper here but can also be done with Conda 
workon pycoMeth
```

## Getting help

```bash
pycoMeth Comp_Report --help
```

## Example usage

#### Example with a single significant result

```bash
pycoMeth Comp_Report \
    -i "./data/Yeast_CGI_meth_comp.tsv.gz" \
    -g "./data/yeast.gff3" \
    -o "yeast_html" \
    --pvalue_threshold 0.05 \
    --verbose
```

#### Usage with large dataset

```bash
pycoMeth Comp_Report \
    -i "./data/Medaka_CGI_meth_comp.tsv.gz" \
    -g "./data/medaka.gff3" \
    -o "medaka_html" \
    --n_top 100 \
    --progress
```
