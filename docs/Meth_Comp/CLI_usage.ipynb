# Meth_Comp CLI usage

## Activate virtual environment

```bash
# Using virtualenvwrapper here but can also be done with Conda 
workon pycoMeth
```

## Getting help

```bash
pycoMeth Meth_Comp --help
```

## Example usage

#### Usage with CpG Aggregate output

```bash
pycoMeth Meth_Comp \
    -i "./data/Yeast_sample_1_CpG.tsv.gz" "./data/Yeast_sample_2_CpG.tsv.gz" "./data/Yeast_sample_3_CpG.tsv.gz" "./data/Yeast_sample_4_CpG.tsv.gz" \
    -f "./data/yeast.fa" \
    -b "./results/CLI_CpG_Yeast.bed" \
    -t "./results/CLI_CpG_Yeast.tsv" \
    -s S1 S2 S3 S4 \
    -l 0.5 \
    --progress

head ./results/CLI_CpG_Yeast.bed
head ./results/CLI_CpG_Yeast.tsv
```

#### Usage with Interval Aggregate output

```bash
pycoMeth Meth_Comp \
    -i "./data/Medaka_sample_1_CGI.tsv.gz" "./data/Medaka_sample_2_CGI.tsv.gz" "./data/Medaka_sample_3_CGI.tsv.gz" "./data/Medaka_sample_4_CGI.tsv.gz" \
    -f "./data/medaka.fa" \
    -b "./results/CLI_Interval_Medaka.bed" \
    -t "./results/CLI_Interval_Medaka.tsv" \
    -s S1 S2 S3 S4 \
    --progress

head ./results/CLI_Interval_Medaka.bed
head ./results/CLI_Interval_Medaka.tsv
```
